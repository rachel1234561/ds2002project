```python
#Link: https://www.socialexplorer.com/tables/ACS2022_5yr/R13772573
earnings_edu = pd.read_csv("earnings_edu.csv")
earnings_edu=earnings_edu.rename(columns={'SE_A15001_001':'Median Earnings',"SE_A15001_002":"Less than high school","SE_A15001_003":"High school","SE_A15001_004":"Some college or Associates","SE_A15001_005":"Bachelors","SE_A15001_006":"Graduates or Professional","Geo_QName":"State"})
earnings_edu=earnings_edu[['State','Median Earnings','Less than high school','High school','Some college or Associates','Bachelors','Graduates or Professional']]
earnings_edu.head()
earnings_edu.to_csv('earnings.csv',index=False)
```

```python
edu_stats=pd.read_csv('socialexplorer.csv')
#Link:https://www.socialexplorer.com/tables/ACS2023/R13770060
cols=edu_stats.columns
cols2=list(cols[2:53])
edu_stats.drop(cols2,axis='columns',inplace=True)
edu_stats.drop('Geo_FIPS',axis='columns',inplace=True)
edu_stats.drop('SE_A14024_001',axis='columns',inplace=True)
edu_stats=edu_stats.rename(columns={"PCT_SE_A12002_002":"% Less than high school","PCT_SE_A12002_003":" % High school","PCT_SE_A12002_004":"% Some college","PCT_SE_A12002_005":"% Bachelors","PCT_SE_A12002_006":"% Masters","PCT_SE_A12002_007":"% Professional degree","PCT_SE_A12002_008":"% Doctorate","Geo_QName":"State"})
edu_stats
```

```python
inst_lvl=pd.read_csv('Institution-Level.csv')
#Link:https://collegescorecard.ed.gov/data/
inst_lvl=inst_lvl[['INSTNM','STABBR','ICLEVEL','NPT4_PUB','NPT4_PRIV','UGDS','MD_EARN_WNE_5YR']]
inst_lvl.fillna({'NPT4_PRIV': 0}, inplace=True)
inst_lvl.fillna({'NPT4_PUB': 0}, inplace=True)
inst_lvl['AVG_Net_Price']=inst_lvl['NPT4_PUB']+inst_lvl['NPT4_PRIV']
inst_lvl.drop(['NPT4_PUB','NPT4_PRIV'], axis=1, inplace=True)
inst_lvl.drop(inst_lvl[inst_lvl['AVG_Net_Price']==0].index,inplace=True)
inst_lvl.dropna(inplace=True)
inst_lvl.replace({'ICLEVEL': {1: '4 Year',2: '2 Year',3: '<2 Year'}}, inplace=True)
inst_lvl=inst_lvl.rename(columns={'MD_EARN_WNE_5YR':'MD_5YR_SALARY','ICLEVEL':'INST_TYPE','STABBR':'STATE','UGDS':'UNDG_POP','INSTNM':'Name'})
inst_lvl['ROI_overall']=(inst_lvl['MD_5YR_SALARY']/inst_lvl['AVG_Net_Price'])*100
inst_lvl=inst_lvl[['Name','STATE','MD_5YR_SALARY','AVG_Net_Price']]
inst_lvl.to_csv('institution.csv',index=False)
```
