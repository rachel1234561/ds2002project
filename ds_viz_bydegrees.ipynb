```python
import pandas as pd
earnings_edu = pd.read_csv("earnings_edu.csv")
earnings_edu=earnings_edu.rename(columns={'SE_A15001_001':'Median Earnings',"SE_A15001_002":"Less than high school","SE_A15001_003":"High school","SE_A15001_004":"Some college or Associates","SE_A15001_005":"Bachelors","SE_A15001_006":"Graduates or Professional","Geo_QName":"State"})
earnings_edu = earnings_edu[['State','Median Earnings','Less than high school','High school','Some college or Associates','Bachelors','Graduates or Professional']]
earnings_edu.head()
```

```python
import matplotlib.pyplot as plt
```

```python
earnings_edu.set_index('State', inplace=True)
```

```python
earnings_edu_no_median = earnings_edu.drop(columns=['Median Earnings'])
transposed = earnings_edu_no_median.transpose()
max_grad_state = earnings_edu['Graduates or Professional'].idxmax()
min_grad_state = earnings_edu['Graduates or Professional'].idxmin()

plt.figure(figsize=(15, 8))
for state in transposed.columns:
    plt.plot(transposed.index, transposed[state], label=state)

max_grad_earnings = earnings_edu.loc[max_grad_state, 'Graduates or Professional']
min_grad_earnings = earnings_edu.loc[min_grad_state, 'Graduates or Professional']
plt.annotate(max_grad_state, 
             xy=('Graduates or Professional', max_grad_earnings), 
             xytext=('Graduates or Professional', max_grad_earnings + 10000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=12, weight='bold')
plt.annotate(min_grad_state, 
             xy=('Graduates or Professional', min_grad_earnings), 
             xytext=('Graduates or Professional', min_grad_earnings - 10000),
             arrowprops=dict(facecolor='orange', shrink=0.05),
             fontsize=12, weight='bold')

## plt.figure(figsize=(15, 8))
##for state in transposed.columns:
    #plt.plot(transposed.index, transposed[state], label=state)
plt.title('Earnings by Education Level for Each State', fontsize = 20, weight = "bold", pad = 20)
plt.xlabel('Education Level', fontsize = 15, weight = "bold", labelpad = 20)
plt.ylabel('Earnings', fontsize = 15, weight = "bold", labelpad = 20)
plt.xticks(fontsize = 12, weight = "bold")
plt.yticks(fontsize = 12)
#plt.legend(title='State', loc='upper left', bbox_to_anchor=(1, 1))
plt.gca().tick_params(axis='x', pad=10)
plt.tight_layout()
plt.show()
```
